```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from scipy import stats
plt.style.use('ggplot')
```

```python
rng = np.random.default_rng(77)
```

```python
mu1 = 25
mu2 = 10
cov_xy = 5
cov_yx = 5
var_x = 9
var_y = 4


mean = [mu1, mu2]
cov_matrix = [[var_x,cov_xy],
              [cov_yx, var_y]] 
x, y = rng.multivariate_normal(mean, cov= cov_matrix, size = 1000).T
```

### Covariance matrix

```python
np.cov(x,y)
```

### Perason correlation coefficient

```python
np.corrcoef(x,y)
```

```python
plt.figure(figsize = (10,10))

plt.subplot(2,2,2)
plt.scatter(x,y)
plt.title('Joint distribution of X and Y')

plt.subplot(2,2,1)
plt.hist(y, bins = 15, orientation = 'horizontal', ec = 'black' )
plt.title('Marginal Distribution of Y')

plt.subplot(2,2,4)
plt.hist(x, bins = 15, ec = 'black')
plt.title('Marginal Distribution of X')
```

### Plotting bivariate distribution

```python
data = rng.multivariate_normal(mean, cov= cov_matrix, size = 1000)

df = pd.DataFrame(data, columns = ['x','y'])
```

```python
j = sns.jointplot(x='x', y='y', data = df, kind = 'kde', height = 8)
j.annotate(stats.pearsonr)
```
