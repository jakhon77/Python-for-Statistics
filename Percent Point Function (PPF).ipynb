### PDF: Probability Density Function, returns the probability of a given continuous outcome.
### CDF: Cumulative Distribution Function, returns the probability of a value less than or equal to a given outcome.
### PPF: Percent-Point Function, returns a discrete value that is less than or equal to the given probability.

```python
import numpy as np
from scipy.stats import norm
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mu = 45
std = 5
dist = norm(mu, std)
```

```python
values = np.linspace(dist.mean()-4*dist.std(), dist.mean()+4*dist.std(), 100)
```

# PDF

```python
probabilities = dist.pdf(values)
```

```python
plt.plot(values,probabilities)
plt.title('PDF of the function')
plt.xlabel('Values')
plt.ylabel('Probability')
```

# CDF

```python
c_probabilities = dist.cdf(values)
```

```python
plt.plot(values,c_probabilities)
plt.title('CDF of the function')
plt.xlabel('Values')
plt.ylabel('Probability')
```

# PPF

```python
fig = plt.figure(figsize = (15,8))

plt.subplot(121)
plt.plot(c_probabilities,values)
plt.title('PPF of the function (by flipping CDF)')
plt.xlabel('Probability')
plt.ylabel('Values')

plt.subplot(122)
probs = np.linspace(0,1,100)
plt.plot(probs,dist.ppf(probs))
plt.title('PPF of the function (constructed manually)')
plt.xlabel('Probability')
plt.ylabel('Values')
```

```python
lower_end = dist.ppf(0.025)
upper_end = dist.ppf(0.975)
print(f"Middle 95% lies between {lower_end.round(2)} and {upper_end.round(2)}")

# Let's check if it is true
dist.cdf(54.8)-dist.cdf(35.2)
```
