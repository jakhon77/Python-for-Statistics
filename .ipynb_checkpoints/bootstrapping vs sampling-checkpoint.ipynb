```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
# let's create random number generator
rng = np.random.default_rng(77)
```

```python
# population of UofA
all_students = 33000
```

```python
# first partition of the population
mu1 = 180
std1 = 6
gym_goers = rng.normal(mu1, std1, size = int(all_students*0.4))
len(gym_goers)
```

```python
# second partition of the population
mu2 = 150
std2 = 6
non_gym_goers = rng.normal(mu2, std2, size = int(all_students - all_students*0.4 ))
len(non_gym_goers)
```

```python
# the entire population
population = np.hstack((gym_goers, non_gym_goers))
len(population)
```

### Let's visualize the distribution of the entire population, first partition and second partition

```python
plt.figure (figsize = (15,12)) 

plt.subplot(311)
sns.distplot(population)
plt.xlim(120,200)
plt.title("Population distribution")
plt.axvline(x = np.mean(population), color = 'red')

plt.subplot(312)
sns.distplot(gym_goers)
plt.title("Gym_goers distribution")
plt.axvline(x = np.mean(gym_goers), color = 'red')
plt.xlim(120,200)

plt.subplot(313)
sns.distplot(non_gym_goers)
plt.title("Non-gym goers distribution")
plt.axvline(x = np.mean(non_gym_goers), color = 'red')
plt.xlim(120,200)
```

#### First sample from the entire population. Then, we will sample using bootstraping method 

```python
n_samples = 1000
sample_size = 100

# let's sample from the population itself
sample_means = [np.mean(rng.choice(population, size = 100)) for _ in range(n_samples)]

plt.figure (figsize = (15,12)) 

plt.subplot(311)
sns.distplot(population)
plt.xlim(120,200)
plt.title(f"Population distribution, mu = {np.mean(population).round(2)}")
plt.axvline(x = np.mean(population), color = 'red')

plt.subplot(312)
sns.distplot(sample_means)
plt.xlim(120,200)
plt.title(f"Sample means distribution, mu = {np.mean(sample_means).round(2)}")
plt.axvline(x = np.mean(sample_means), color = 'red')

# Now, let's sample using bootstraping method

single_sample = rng.choice(population, size = 100)
boot_sample_means = [np.mean(rng.choice(single_sample, size = 100)) for _ in range(n_samples)]

plt.subplot(313)
sns.distplot(boot_sample_means)
plt.xlim(120,200)
plt.title(f"Bootstrapped sample means distribution, mu = {np.mean(boot_sample_means).round(2)}")
plt.axvline(x = np.mean(boot_sample_means), color = 'red')
```

### As we can see from the graphs above BOOTSTRAPPING method works quite well to identify the mean of the population.
